```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from finrl.meta.preprocessor.preprocessors import FeatureEngineer
from finrl.meta.env_stock_trading.env_stocktrading import StockTradingEnv
from finrl.agents.stablebaselines3.models import DRLAgent
from finrl.config import INDICATORS
import os

# 1. Load and Prepare Apple Data
df_apple = pd.read_pickle("apple_data.pkl")
if df_apple.index.name != 'date':
    df_apple.index.name = 'date'
df_apple.reset_index(inplace=True)
df_apple['tic'] = 'AAPL'  # Apple-only data

# 2. Feature Engineering
fe = FeatureEngineer(use_technical_indicator=True, use_turbulence=False, user_defined_feature=False)
df_apple = fe.preprocess_data(df_apple)
df_apple.set_index('date', inplace=True)

# 3. Add Covariance and Returns
def add_cov_and_returns(df, lookback=252):
    df = df.reset_index().sort_values(['date', 'tic'])
    df.index = df['date'].factorize()[0]
    cov_list = []
    return_list = []
    for i in range(lookback, len(df.index.unique())):
        data_lookback = df.loc[i - lookback:i]
        price_lookback = data_lookback.pivot_table(index='date', columns='tic', values='close')
        return_lookback = price_lookback.pct_change().dropna()
        return_list.append(return_lookback)
        covs = return_lookback.cov().values
        cov_list.append(covs)
    df_cov = pd.DataFrame({
        'date': df['date'].unique()[lookback:],
        'cov_list': cov_list,
        'return_list': return_list
    })
    df_cov.set_index('date', inplace=True)
    df = df.set_index('date').merge(df_cov, how='left', left_index=True, right_index=True)
    return df.reset_index().set_index('date')

df_apple = add_cov_and_returns(df_apple)
df_apple = df_apple.fillna(0)

# 4. Train/Test Split
TRADE_START_DATE = '2023-11-01'
train_df = df_apple[df_apple.index < TRADE_START_DATE]
test_df = df_apple[df_apple.index >= TRADE_START_DATE]

# 5. Environment Setup
def get_env(df):
    stock_dim = 1
    state_space = 1 + 2 * stock_dim + len(INDICATORS) * stock_dim
    env_kwargs = {
        "hmax": 100,
        "initial_amount": 1_000_000,
        "num_stock_shares": [0],
        "buy_cost_pct": [0.001],
        "sell_cost_pct": [0.001],
        "state_space": state_space,
        "stock_dim": stock_dim,
        "tech_indicator_list": INDICATORS,
        "action_space": stock_dim,
        "reward_scaling": 1e-4
    }
    return StockTradingEnv(df=df.reset_index(), **env_kwargs)
```

```python
# 6. Train PPO Agent
env_train = get_env(train_df)
env_train, _ = env_train.get_sb_env()
agent = DRLAgent(env=env_train)
model = agent.get_model("ppo")
trained_model = agent.train_model(model=model, tb_log_name="ppo_apple", total_timesteps=30000)
```

```python
# 7. Evaluate on Test Set
test_env = get_env(test_df)
df_account_value, df_actions = DRLAgent.DRL_prediction(model=trained_model, environment=test_env)

# 8. Prepare Data for Plotting
df_account_value['date'] = pd.to_datetime(df_account_value['date'])
df_account_value.set_index('date', inplace=True)
df_actions = pd.DataFrame(df_actions, columns=['actions'], index=df_account_value.index)
df_test_price = test_df[test_df['tic'] == 'AAPL'][['close']].drop_duplicates()
merged_df = df_account_value.merge(df_test_price, left_index=True, right_index=True, how='left')

# 9. Plot Net Worth, Price, and Actions
plt.figure(figsize=(15, 10))

# Net Worth
plt.subplot(3, 1, 1)
plt.plot(merged_df.index, merged_df['account_value'], color='blue', label='Net Worth')
plt.legend()
plt.ylabel("Net Worth")

# Close Price
plt.subplot(3, 1, 2)
plt.plot(merged_df.index, merged_df['close'], color='orange', label='Close Price')
plt.legend()
plt.ylabel("Close Price")

# Buy/Sell Actions
plt.subplot(3, 1, 3)
buy_actions = df_actions[df_actions['actions'] > 0]
sell_actions = df_actions[df_actions['actions'] < 0]
plt.bar(buy_actions.index, buy_actions['actions'], color='green', label='Buy')
plt.bar(sell_actions.index, sell_actions['actions'], color='red', label='Sell')
plt.legend()
plt.ylabel("Action (Shares)")
plt.xlabel("Date")

plt.tight_layout()
os.makedirs("results", exist_ok=True)
plt.savefig("results/apple_trading_plot.png")
plt.show()

# 10. Performance Metrics

# Sharpe Ratio
account_values = merged_df['account_value']
returns = account_values.pct_change().dropna()
sharpe_ratio = (np.mean(returns) / np.std(returns)) * np.sqrt(252)

# Maximum Drawdown
cumulative_returns = (1 + returns).cumprod()
peak = cumulative_returns.cummax()
drawdown = (cumulative_returns - peak) / peak
max_drawdown = drawdown.min()

print("\n==== Performance Metrics ====")
print(f"Sharpe Ratio: {sharpe_ratio:.4f}")
print(f"Max Drawdown: {max_drawdown:.4%}")
```
