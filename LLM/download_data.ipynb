```python
import yfinance as yf

aapl_df = yf.download('AAPL',
                      progress=False,
)
aapl_df.head()
```

```python
# Copy the original DataFrame
aapl_df_clean = aapl_df.copy()

# Rename the columns
aapl_df_clean.columns = ['close', 'high', 'low', 'open', 'volume']

# Display the cleaned DataFrame
aapl_df_clean.head()
```

```python
aapl_df_clean.to_pickle("apple_data.pkl")
```
